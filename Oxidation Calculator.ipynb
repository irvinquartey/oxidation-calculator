```python
import tkinter as tk
from tkinter import ttk
from tkinter import *
import math
import tkinter.font as tkFont



LARGEFONT =("Verdana", 25)

XLARGEFONT =("Verdana",12)


def temperaturelimit():
    
        if tcheck< 700 or tcheck > 1200:
            
                       
            clear_alli()      
            
            oxi1.insert(10,"Wrong temperature input")

def temperaturelimit1():
    
        if tcheck1< 700 or tcheck1 > 1200:
                    
            clear_all()      
            
            thick.insert(10,"Wrong temperature input") 

    

def clear_all() :

        # This function activates the clearing of all entry boxes for user input

        t1.delete(0, END) 
        t.delete(0, END)
        timeused.delete(0, END)
        thick.delete(0, END)
        clicked1.set( "Select crystal orientation" ) 
        clicked2.set( "Select type of oxidation" ) 

        # Used to set focus on the principle_field entry box 
        t1.focus_set()




def clear_alli() :


        ti1.delete(0, END) 
        ti2.delete(0, END) 
        ti.delete(0, END)
        oxi1.delete(0, END)
        clicked1i.set( "Select crystal orientation" ) 
        clicked2i.set( "Select type of oxidation" ) 

        t1.focus_set()
        
def clearcurrent1():

         oxi1.delete(0, END)
         oxidationtimecalc()
         temperaturelimit()

def clearcurrent2():

         thick.delete(0, END)
         calculateoxide()
         temperaturelimit1()




def oxidationtimecalc():

        global tcheck

        ft=float(ti2.get())
        it=float(ti1.get())
        k = 8.6173*10**-5  #Boltzman's constant in ev/K
        C1weti = 3.86*10**2 
        C2wet111i=1.63*10**8
        C1dryi=7.72*10**2
        C2dry111i=6.23 * 10**6
        C2wet100i=9.7*10**7
        C2dry100i=3.71*10**6
        E1weti=0.78
        E2weti=2.05
        E1dryi=1.23
        E2dryi=2.00
        tcheck=int(ti.get())
        tempi= int(ti.get()) + 273



          
        if clicked2i.get()=="wet":
         E1i=E1weti
         E2i=E2weti
         C1i=C1weti
         if clicked1i.get()=="100":
          C2i = C2wet100i
         else:
          C2i = C2wet111i

        else:
         E1i=E1dryi
         E2i=E2dryi
         C1i=C1dryi
         if clicked1i.get()=="100":
          C2i = C2dry100i
         else:
          C2i = C2dry111i

    
        kT=k*tempi

        Bi= C1i*math.exp(-(E1i)/kT)
        B_Ai= C2i*math.exp(-(E2i)/kT)
        Ai=Bi/B_Ai
        print('Parabolic rate constant=',Bi)
        print('Linear rate constant=',B_Ai)
        

        time = float((ft**2)-(it**2))/Bi + (ft-it)/(B_Ai)
        oxi1.insert(10, round(time,2))

        return






def calculateoxide():  # Function for calculating oxide thickness


        global tcheck1

        d0=float(t1.get())
        k = 8.6173*10**-5  #Boltzman's constant in ev/K
        C1wet = 3.86*10**2 
        C2wet111=1.63*10**8
        C1dry=7.72*10**2
        C2dry111=6.23 * 10**6
        C2wet100=9.7*10**7
        C2dry100=3.71*10**6
        E1wet=0.78
        E2wet=2.05
        E1dry=1.23
        E2dry=2.00
        tcheck1=int(t.get())
        temp= int(t.get()) + 273






        if clicked2.get()=="wet":
         E1=E1wet
         E2=E2wet
         C1=C1wet
         if clicked1.get()=="100":
          C2 = C2wet100
         else:
          C2 = C2wet111

        else:
         E1=E1dry
         E2=E2dry
         C1=C1dry
         if clicked1.get()=="100":
          C2 = C2dry100
         else:
          C2 = C2dry111


    
        kT=k*temp

        B= C1*math.exp(-(E1)/kT)
        B_A= C2*math.exp(-(E2)/kT)
        A=B/B_A

        r=float(1+((4*B/(A*A))*(float(timeused.get()) +((d0*d0)+(A*d0))/B)))
        x1 = float(A/2)*(( math. sqrt(r))-1)
        thick.insert(10, round(x1,2))


        return



def quit(self):
        self.destroy()

       

class tkinterApp(tk.Tk):
        
        # __init__ function for class tkinterApp 
        def __init__(self, *args, **kwargs): 
                
                # __init__ function for class Tk
                tk.Tk.__init__(self, *args, **kwargs)
                
                

                self.title("Oxidation Calculator")
                self.resizable(True, True)
               
                
                # creating a container
                container = tk.Frame(self) 
                container.pack(side = "top", fill = "both", expand = True) 

                container.grid_rowconfigure(0, weight = 1)
                container.grid_columnconfigure(0, weight = 1)



                

                # Set the configuration of GUI window 
        

                # initializing frames to an empty array
                self.frames = {} 

                # iterating through a tuple consisting
                # of the different page layouts
                for F in (StartPage, Page1, Page2):

                        frame = F(container, self)

                        # initializing frame of that object from
                        # startpage, page1, page2 respectively with 
                        # for loop
                        self.frames[F] = frame 

                        frame.grid(row = 0, column = 0, sticky ="nsew")

                self.show_frame(StartPage)

        # to display the current frame passed as
        # parameter
        def show_frame(self, cont):
                frame = self.frames[cont]
                frame.tkraise()

# first window frame startpage

class StartPage(tk.Frame):
        def __init__(self, parent, controller): 
                tk.Frame.__init__(self, parent)

                

                
                # label of frame Layout 2
                label = ttk.Label(self, text ="Oxidation Calculator", font = LARGEFONT)
                
                # putting the grid in its place by using
                # grid
                label.pack()
                #label.grid(row = 0, column = 1, padx = 10, pady = 10)


                label = ttk.Label(self, text ="Please select your option", font = XLARGEFONT)
                
                # putting the grid in its place by using
                # grid
                label.pack()
                #label.grid(row = 1, column = 1, padx = 10, pady = 10) 
                my_font=('verdana', 20, 'bold')


                button1 = tk.Button(self, text ="Oxidation Time Calculator",font=my_font,bg="green",fg="white",
                command = lambda : controller.show_frame(Page1))
        
                # putting the button in its place by
                # using grid
                button1.pack()
                #button1.grid(row = 3, column = 1, padx = 10, pady = 10)
               

                ## button to show frame 2 with text layout2
                button2 = tk.Button(self, text ="Oxide Thickness Calculator",font=my_font,bg="brown",fg="white",
                command = lambda : controller.show_frame(Page2))
                button2.pack()
                
        

                


# second window frame page1 
class Page1(tk.Frame):
        
        def __init__(self, parent, controller):
                tk.Frame.__init__(self, parent)
                label = ttk.Label(self, text ="Oxidation Time Calculation Page", font = LARGEFONT)
                label.grid(row = 0, column = 1, padx = 10, pady = 10)
                
                # button to show frame 2 with text
                # layout2

                my_font=('verdana', 11, 'bold')
                label10 = Label(self, text = "Navigation options ",font=my_font, 
                                fg = 'white', bg = 'green') 

 
                label10.grid(row = 10, column = 0, padx = 10, pady = 10)


               
                button1 = Button(self, text ="Calc StartPage",font=my_font,command = lambda : controller.show_frame(StartPage))
        
                # putting the button in its place 
                # by using grid
                button1.grid(row = 11, column = 0, padx = 10, pady = 10)

                # button to show frame 2 with text
                # layout2
                
                button2 = Button(self, text ="Thickness Calc Page",font=my_font,
                                                        command = lambda : controller.show_frame(Page2))
        
                # putting the button in its place by 
                # using grid
                button2.grid(row = 12, column = 0, padx = 10, pady = 10)
                




                global clicked2i
                global clicked1i


                # Dropdown menu options 
                options1 = ["100","111"]
                # datatype of menu text 
                clicked1i = StringVar()

                click1i = clicked1i.get()

                # initial menu text 
                clicked1i.set( "Select crystal orientation" ) 
                
                my_font=('verdana', 13, 'bold')
                

                # Create Dropdown menu 
                drop1 = OptionMenu( self , clicked1i , *options1 ) 
                drop1.config(font=my_font)
                drop1.grid(row = 1, column = 0, padx = 10, pady = 10)


                # Change the label text 
                #def show(): 
                #label.config( text = clicked.get() ) 

                # Dropdown menu options 
                options2 = ["wet","dry"]
                # datatype of menu text 
                clicked2i = StringVar()

                click2i = clicked2i.get()

                # initial menu text 
                clicked2i.set( "Select type of oxidation" ) 
                
                


                # Create Dropdown menu 
                drop2 = OptionMenu( self , clicked2i , *options2 ) 
                drop2.config(font=my_font)
                drop2.grid(row = 1, column = 1, padx = 10, pady = 10)

        
                #end of drop down options
                my_font=('verdana', 12, 'bold')

                # Create an Initial thickness : label 
                label1 = Label(self, text = "Initial  thickness (in  micrometers)",font= my_font,
                                fg = 'black', bg = 'white') 

                # Create a Final thickness : label 
                label2 = Label(self, text = "Desired thickness(in micrometers)", font= my_font,
                                fg = 'black', bg = 'white') 
        
                # Create a Temperature : label 
                label3 = Label(self, text = "Temperature(in °C)(Keep between 700 and 1200) ", font= my_font,
                                fg = 'black' , bg = 'white')

                # Create an Oxidation time : label 
                label4 = Label(self, text = "     Oxidation time( in Hours)       ", font= my_font,
                                fg = 'black', bg = 'white')
 
           

                # grid method is used for placing 
                # the widgets at respective positions 
                # in table like structure .
                my_font3=('verdana', 11, 'bold')
                # padx keyword argument used to set paading along x-axis .
                # pady keyword argument used to set paading along y-axis . 
                label1.grid(row = 3, column = 0, padx = 10, pady = 10) 
                label2.grid(row = 5, column = 0, padx = 10, pady = 10) 
                label3.grid(row = 6, column = 0, padx = 10, pady = 10)
                label4.grid(row = 8, column = 0, padx = 10, pady = 10)

                global ti1
                global ti2
                global ti
                global oxi1

                ti1 = Entry(self, font = XLARGEFONT) 
                ti2 = Entry(self, font = XLARGEFONT) 
                ti = Entry(self, font = XLARGEFONT)
                oxi1 = Entry(self, font = my_font3)
                
                
                ti1.grid(row = 3, column = 1, padx = 10, pady = 10) 
                ti2.grid(row = 5, column = 1, padx = 10, pady = 10) 
                ti.grid(row = 6, column = 1, padx = 10, pady = 10)
                oxi1.grid(row = 8, column = 1, padx = 10, pady = 10)
              
                
                my_font=('verdana', 15, 'bold')

                button3 = Button(self, text = "Calculate oxidation time", bg = "green",font= my_font,
                                        fg = "white", command = clearcurrent1) 

                button4 = Button(self, text = "Clear", bg = "brown", font= my_font ,
                                        fg = "white", command = clear_alli)

                button3.grid(row = 10, column = 1, pady = 10) 
                button4.grid(row = 11, column = 1, pady = 10)




class Page2(tk.Frame): 
        def __init__(self, parent, controller):
                tk.Frame.__init__(self, parent)
                label = ttk.Label(self, text ="Oxide Thickness Calculation Page", font = LARGEFONT)
                label.grid(row = 0, column = 1, padx = 10, pady = 10)

                # button to show frame 2 with text
                
                my_font=('verdana', 11, 'bold')
                # layout2

                label10 = Label(self, text = "Navigation options ", font=my_font,

                                fg = 'white', bg = 'green') 

 
                label10.grid(row = 10, column = 0, padx = 10, pady = 10)        


                
                button1 = Button(self, text ="Calc StartPage", font=my_font,
                                                        command = lambda : controller.show_frame(StartPage))
        
                button1.grid(row = 11, column = 0, padx = 10, pady = 10)

                # button to show frame 2 with text
                # layout2
                
                button2 = Button(self, text ="Oxidation Time Calc Page",font=my_font,
                                                        command = lambda : controller.show_frame(Page1))
        
                # putting the button in its place by 
                # using grid
                button2.grid(row = 12, column = 0, padx = 10, pady = 10)
                
   


                global clicked2
                global clicked1

                # Dropdown menu options 
                options1 = ["100","111"]
                # datatype of menu text 
                clicked1 = StringVar()

                click1 = clicked1.get()

                # initial menu text 
                clicked1.set( "Select crystal orientation" ) 
                my_font=('verdana', 13, 'bold')
                

                # Create Dropdown menu 
                drop1 = OptionMenu( self , clicked1 , *options1 )
                drop1.config(font=my_font)
                drop1.grid(row = 1, column = 0, padx = 10, pady = 10)


                my_font=('verdana', 13, 'bold')
                

                # Dropdown menu options 
                options2 = ["wet","dry"]
                # datatype of menu text 
                clicked2 = StringVar()

                click2 = clicked2.get()

                # initial menu text 
                clicked2.set( "Select type of oxidation" ) 

                # Create Dropdown menu 
                drop2 = OptionMenu( self , clicked2 , *options2 ) 
                drop2.config(font=my_font)
                drop2.grid(row = 1, column = 1, padx = 10, pady = 10)

        
                #end of drop down options
                my_font=('verdana', 12, 'bold')

                # Create an Initial thickness : label 
                label1 = Label(self, text = "Initial thickness (in micrometers)", font= my_font ,
                                fg = 'black', bg = 'white') 

        
                # Create a Temperature : label 
                label3 = Label(self, text = "Temperature(in °C)(Keep between 700 and 1200) ",   font= my_font ,
                                fg = 'black', bg = 'white')

 
                # Create a time used : label 
                label5 = Label(self, text = "     Time Used(in Hours)            ",  font= my_font , 
                                fg = 'black', bg = 'white')

                # Create a time used : label 
                label6 = Label(self, text = "Oxide Thickness(micrometers) ",   font= my_font ,
                                fg = 'black', bg = 'white') 

                # grid method is used for placing 
                # the widgets at respective positions 
                # in table like structure .

                # padx keyword argument used to set paading along x-axis .
                # pady keyword argument used to set paading along y-axis . 
                label1.grid(row = 3, column = 0, padx = 10, pady = 10) 
                label3.grid(row = 5, column = 0, padx = 10, pady = 10)
                label5.grid(row = 6, column = 0, padx = 10, pady = 10)
                label6.grid(row = 8, column = 0, padx = 10, pady = 10)

                # Create an entry box for filling or typing the information.



                global t1
                global t
                global timeused
                global thick

                my_font3=('verdana', 11, 'bold')
                
                t1 = Entry(self, font = XLARGEFONT) 
                t = Entry(self, font = XLARGEFONT)
                timeused  = Entry(self, font = XLARGEFONT)
                thick = Entry(self, font = my_font3)


                t1.grid(row = 3, column = 1, padx = 10, pady = 10)  
                t.grid(row = 5, column = 1, padx = 10, pady = 10)
                timeused.grid(row = 6, column = 1, padx = 10, pady = 10)
                thick.grid(row = 8, column = 1, padx = 10, pady = 10)
        
                my_font=('verdana', 15, 'bold')
                # Create a Button  for calculation after entering values
    
                button3 = Button(self, text = "Calculate oxide thickness", bg = "green", font=my_font,
                                        fg = "white", command = clearcurrent2) 

                # Create a Clear Button to reset values
     
                button4 = Button(self, text = "Clear", bg = "brown",font=my_font,
                                        fg = "white", command = clear_all)

                button3.grid(row = 10, column = 1, pady = 10) 
                button4.grid(row = 11, column = 1, pady = 10)

# Driver Code
app = tkinterApp()
app.mainloop()
```
